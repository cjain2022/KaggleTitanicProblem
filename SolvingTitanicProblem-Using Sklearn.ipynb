```python
#PassengerId - Not Deciding factor for person survival  
#Survived - Outcome[Y]       
#Pclass -Deciding Factor
    #pclass: A proxy for socio-economic status (SES)
    #1st = Upper
    #2nd = Middle
    #3rd = Lower        
#Name -Not Deciding factor for person survival          
#Sex- May be deciding             
#Age -May be deciding            
#SibSp -May be Deciding Factor
    # No.of siblings / spouses aboard the Titanic
#Parch -May be deciding Factor
    # No. of parents / children aboard the Titanic
#Ticket -Ticket Number cannot be deciding factor        
#Fare- Fare can be deciding factor depending upon if it was business class ticket etc         
#Cabin - Not Deciding factor for person survival
#Embarked - Not Deciding factor for person survival
    #Port of Embarkation
    #C = Cherbourg, Q = Queenstown, S = Southampton
```

```python
cols_to_remove=['PassengerId','Name','Ticket','Embarked','Cabin']
```

```python
clean_data=data.drop(labels=cols_to_remove, axis=1)
clean_data.head()
clean_data.info()
```

```python
# Now We need to convert Sex into numeric data 
#sex_num={'male':1, 'female':0}
#clean_data['Sex'].size
#for i in range(clean_data['Sex'].size):
    #clean_data['Sex'][i]=sex_num[clean_data['Sex'][i]]
#Option 2 Use Sklearn  
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
clean_data['Sex']=le.fit_transform(data['Sex'])
```

```python
# Age Missing Entries Fill Up 
clean_data=clean_data.fillna(value=int(clean_data['Age'].mean()))
clean_data.info()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer 
imp_mean=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
clean_data.info()
```

```python
#To View Any Row Details 
clean_data.loc[2]
```

```python
#To Separate X and Y 
Xcols=['Sex','Age','SibSp','Parch','Fare','Pclass']
Ycols=['Survived']
Xdata=clean_data.drop(Ycols, axis=1)
Y=clean_data.drop(Xcols, axis=1)

#This Contains x and y both 
#xdata=clean_data
```

```python
Xdata
```

```python
Y
```

```python
#Spliting 
split=int(0.7*Xdata.shape[0])
Xtrain=Xdata[:split]
Ytrain=Y[:split]
Xtest=Xdata[split:]
Xtest.reset_index(drop=True) #Imp
Ytest_actual=Y[split:]
Ytest_actual.reset_index(drop=True) # IMP
```

```python
from sklearn.tree import DecisionTreeClassifier
sk_tree=DecisionTreeClassifier(criterion='entropy',max_depth=5)
sk_tree.fit(Xtrain,Ytrain)
```

```python
y_pred=sk_tree.predict(Xtest)
```

```python
accuracy=sk_tree.score(Xtest,Ytest_actual)
```
