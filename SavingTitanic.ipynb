```python
deletecols=['name','ticket','cabin','embarked','boat','boat','body','home.dest']
xcols=['pclass','sex','age','sibsp','parch','fare']
ycols=['survived']

clean_data=df.drop(deletecols,axis=1)
Ydata=clean_data.drop(xcols,axis=1)
Xdata=clean_data.drop(ycols,axis=1)
```

```python
#Xdata.head()
#Ydata.head()
```

```python
Xdata=Xdata.fillna( Xdata['age'].mean() )
```

```python
Xdata.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Xdata['sex']=le.fit_transform(Xdata['sex'])
```

```python
from sklearn.ensemble import RandomForestClassifier 
rf=RandomForestClassifier(n_estimators=30,criterion='entropy',max_depth=10)
Ylabels=Ydata.columns
ylabel=Ylabels[0]
rf.fit(Xdata,Ydata[ylabel])
rf.score(Xdata,Ydata[ylabel])

from sklearn.model_selection import cross_val_score
acc=cross_val_score( RandomForestClassifier(n_estimators=500,max_depth=10,criterion='entropy'),Xdata,Ydata[ylabel],cv=5).mean()
print(acc)
```
